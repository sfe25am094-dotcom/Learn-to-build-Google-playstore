```python
import pandas as pd
import plotly.express as px
from datetime import datetime, time

# -------- TIME CONDITION (6PMâ€“8PM IST) --------
current_time = datetime.now().time()
if not (time(18,0) <= current_time <= time(20,0)):
    print("Graph visible only between 6 PM to 8 PM IST")
    exit()

# -------- SAMPLE DATA --------
data = {
    "Country":["India","USA","Germany","Brazil","Canada","France","Japan"],
    "Category":["Business","Education","Health","Tools","Finance","Beauty","Music"],
    "Installs":[1500000,800000,2000000,1200000,900000,1100000,1300000]
}

df = pd.DataFrame(data)

# -------- FILTER CATEGORY NOT START WITH A,C,G,S --------
df = df[~df["Category"].str.startswith(tuple(["A","C","G","S"]))]

# -------- TOP 5 CATEGORY BY INSTALLS --------
top_cat = df.groupby("Category")["Installs"].sum().nlargest(5).index
df = df[df["Category"].isin(top_cat)]

# -------- HIGHLIGHT INSTALLS >1 MILLION --------
df["Highlight"] = df["Installs"] > 1000000

# -------- CHOROPLETH MAP --------
fig = px.choropleth(
    df,
    locations="Country",
    locationmode="country names",
    color="Installs",
    hover_name="Category",
    title="Global Installs by App Category"
)

fig.show()
```
