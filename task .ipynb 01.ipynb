```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, time

# -------- TIME CONDITION (3PMâ€“5PM IST) --------
current_time = datetime.now().time()
if not (time(15,0) <= current_time <= time(17,0)):
    print("Graph visible only between 3 PM to 5 PM IST")
    exit()

# -------- SAMPLE DATA --------
data = {
    "Category":["Business","Beauty","Education","Games","Tools",
                "Health","Finance","Music","Food","Travel"],
    "Installs":[50000,40000,60000,70000,30000,35000,20000,45000,25000,15000],
    "Rating":[4.5,4.2,4.6,3.9,4.1,4.3,4.0,4.4,4.2,4.1],
    "Reviews":[10000,9000,12000,8000,7000,7500,5000,8500,6000,4000],
    "Size_MB":[15,20,25,12,11,18,14,16,13,10],
    "Last_Update":["Jan","Jan","Feb","Jan","Jan","Jan","Mar","Jan","Jan","Jan"]
}

df = pd.DataFrame(data)

# -------- FILTER CONDITIONS --------
df = df[df["Rating"] >= 4.0]      # avg rating >=4
df = df[df["Size_MB"] >= 10]      # size >=10MB
df = df[df["Last_Update"] == "Jan"]   # last update Jan

# -------- TOP 10 CATEGORY BY INSTALLS --------
top10 = df.sort_values("Installs", ascending=False).head(10)

# -------- GROUPED BAR CHART --------
x = range(len(top10))

plt.figure(figsize=(10,5))
plt.bar([i-0.2 for i in x], top10["Rating"], width=0.4)
plt.bar([i+0.2 for i in x], top10["Reviews"], width=0.4)

plt.xticks(x, top10["Category"])
plt.xlabel("App Category")
plt.ylabel("Values")
plt.title("Average Rating vs Total Reviews (Top Categories)")
plt.show()
```
